# <font color='#2F4F4F'>AfterWork Data Science: Web Scraping with Python</font>

## <font color='#2F4F4F'>Prerequisites</font>

```
# We first import the required libraries
# ---
#
import pandas as pd             # library for data manupation
import requests                 # library for fetching a web page 
from bs4 import BeautifulSoup   # library for extrating contents from a webpage 
```

## <font color='#2F4F4F'>Step 1: Obtaining our Data</font>

```
# PigiaMe: https://www.pigiame.co.ke/it-software-jobs
# ---
#
pigia_me = requests.get('https://www.pigiame.co.ke/it-software-jobs')
pigia_me
```

```
# MyJobMag: https://www.myjobmag.co.ke/jobs-by-field/information-technology
# ---
#
MyJobMag= requests.get('https://www.myjobmag.co.ke/jobs-by-field/information-technology')
MyJobMag
```

```
# KenyanJob: https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133
# ---
#
KenyanJob= requests.get('https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133')
KenyanJob
```

## <font color='#2F4F4F'>Step 2: Parsing</font>

```
# Parsing our document: pigia_me
# ---
# 
pigia_me_soup = BeautifulSoup(pigia_me.text, 'html.parser')
pigia_me_soup
```

```
# Parsing our document: my_job_mag
# ---
#  
MyJobMag_soup = BeautifulSoup(MyJobMag.text, 'html.parser')
MyJobMag_soup
```

```
# Parsing our document: kenyan_job
# ---
# 
KenyanJob_soup = BeautifulSoup(KenyanJob.text, 'html.parser')
KenyanJob_soup
```

## <font color='#2F4F4F'>Step 3: Extracting Required Elements</font>

```
# 1. Extracting job titles and links: pigia me
# ---
# 
pigia_me_tags = pigia_me_soup.find_all('a', {'class': 'listing-card__inner'})
pigia_me_jobdesc=[] 
pigia_me_link= []


for tags in pigia_me_tags:
  jobdesc= tags.get('data-t-listing_title')
  link=tags.get('href')

  pigia_me_jobdesc.append(jobdesc)
  pigia_me_link.append(link)

```

```
# 2. Extracting job titles: my_job_mag
# ---
# 
my_job_mag_tags = MyJobMag_soup.find_all('li', {'class': 'mag-b'})
my_job_mag_jobdesc=[] 
my_job_mag_link= []

for li in my_job_mag_tags:
  mag_jobdescription = li.find('a').text
  mag_link =  li.find("a")["href"]
  
  my_job_mag_jobdesc.append(mag_jobdescription)
  my_job_mag_link.append("https://www.myjobmag.co.ke" + mag_link)


```

```
# 3. Extracting job titles: kenya_job
# ---
#
kenya_job_tags = KenyanJob_soup.find_all('div', {'class': 'col-lg-5 col-md-5 col-sm-5 col-xs-12 job-title'})
kenya_job_jobdesc=[] 
kenya_job_links= []

for h5 in kenya_job_tags:
  kenya_job_desc = h5.find('a').text
  kenya_job_link = h5.find('a')['href']
  
  kenya_job_jobdesc.append(kenya_job_desc)
  kenya_job_links.append('https://www.kenyajob.com' + kenya_job_link)
```

## <font color='#2F4F4F'>Step 4: Saving our Data</font>

```
# Saving the scraped contents in a dataframe and preview our data
# ---
#Create 2D list and save as dataframe with two named columns.

techjobs_list = list(zip(pigia_me_jobdesc+my_job_mag_jobdesc+kenya_job_jobdesc,my_job_mag_link+pigia_me_link+kenya_job_links))
techjobs_list_df = pd.DataFrame(techjobs_list, columns=['Description','Link'])
```
